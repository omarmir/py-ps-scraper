```python
import bs4 as bs
import urllib.request as urllib
import pandas as pd
```

```python
url = "https://www.tbs-sct.gc.ca/pubs_pol/hrpubs/coll_agre/rates-taux-eng.asp"
page = urllib.urlopen(url).read()
soup = bs.BeautifulSoup(page,'lxml')
```

```python
opts = soup.select('select#dropdown option[value]')
ids = []
urls = []
classifications = []

for item in opts:
    ids.append(item.get('value').split('#')[0].split('?id=')[-1])
    urls.append(item.get('value').split('#')[0])
    classifications.append(item.get('label'))

uniqueURLs = set(urls)
uniqueURLs.remove('/pubs_pol/hrpubs/coll_agre/rates-taux-eng.asp')
```

```python
htmlContent = []
for idURL in uniqueURLs:
    htmlContent.append(urllib.urlopen(idURL).read())
```

```python
uniqueIds = [uniqueURL.split('=')[-1] for uniqueURL in uniqueURLs]
uniqueIdsContent = {'ids' : uniqueIds, 'htmlContent' : htmlContent}
idContentDf = pd.DataFrame(uniqueIdsContent)
```

```python
classDict = {'ids' : ids, 'urls' : urls, 'classifications' : classifications}
classDf = pd.DataFrame(classDict)
classDf = classDf.drop([0])
```

```python
classDf.to_pickle('classDf.pkl')
idContentDf.to_pickle('idContentDf.pkl')
```
